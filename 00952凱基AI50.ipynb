# 凱基優選台灣AI 50 ETF基金

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

url = "https://www.kgifund.com.tw/Fund/Detail?fundID=J020"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Find the h4 tag with the text "股票"
h4_tag = soup.find('h4', class_='fund-asset__sub-title', string='股票')

# Find the table that comes after the h4 tag
table = h4_tag.find_next('table')
#print(str(table).replace("display:none",""))
df_all_stocks = pd.read_html(StringIO(str(table).replace("display:none","")))[0]

print("所有股票持股:")
display(df_all_stocks)
```
